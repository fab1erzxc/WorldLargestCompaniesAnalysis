```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\PythonProjects\webscrapping\worlds_largest_companies.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Revenue'].str.replace('$', '').str.replace(
    ',', '').astype(float), bins=20, kde=True)
plt.title('Distribution of Revenue')
plt.xlabel('Revenue ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Revenue', y='Profit')
plt.title('Revenue vs Profit')
plt.xlabel('Revenue ($)')
plt.ylabel('Profit ($)')
plt.show()
```
